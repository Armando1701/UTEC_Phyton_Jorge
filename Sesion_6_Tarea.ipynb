### **Ejercicio 1:** Método gráfico de un dato categórico ✋

Analiza de manera gráfica un dato categórico de tu dataset.

```python
#Tu código aquí
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Obtener la cantidad de pasajeros por año
pasajeros_por_año = df.groupby('year')["passengers"].sum()

# Obtener la cantidad de pasajeros por mes
pasajeros_por_mes = df.groupby('month')["passengers"].sum()


#Creación de gráficos

plt.figure(figsize=(10, 8)) # Define el tamaño del gráfico (ancho, alto)
pasajeros_por_año.plot(kind='bar', color='skyblue') # Crea el gráfico de barras
bars = pasajeros_por_año.plot(kind='bar', color='skyblue')


# Añadir títulos y etiquetas para que el gráfico sea claro
plt.title('Cantidad Total de Pasajeros por Año', fontsize=16)
plt.xlabel('Año', fontsize=14)
plt.ylabel('Cantidad de Pasajeros', fontsize=14)
plt.xticks(rotation=45) # Rota las etiquetas del eje X para que no se superpongan
plt.grid(axis='y', linestyle='--', alpha=0.7) # Añade una cuadrícula suave en el eje Y

for bar in bars.patches:
    # 'bar.get_height()' obtiene la altura de la barra (el valor)
    # 'bar.get_x() + bar.get_width() / 2' calcula la posición X central de la barra
    # 'bar.get_height()' + un pequeño offset para que el texto esté justo encima
    plt.text(bar.get_x() + bar.get_width() / 2, # Posición X del texto
             bar.get_height() + (pasajeros_por_año.max() * 0.02), # Posición Y del texto (un poco por encima de la barra)
             f'{int(bar.get_height()):,}', # El valor a mostrar, formateado como entero con miles
             ha='center', # Alineación horizontal: centrado sobre la barra
             va='bottom', # Alineación vertical: la parte inferior del texto estará en la posición Y
             fontsize=10,
             color='dimgray')

plt.tight_layout() # Ajusta el diseño para evitar recortes

# Mostrar el gráfico
plt.show()

```

```python
plt.figure(figsize=(10, 8)) # Define el tamaño del gráfico (ancho, alto)
pasajeros_por_mes.plot(kind='bar', color='skyblue') # Crea el gráfico de barras
etiquetas = pasajeros_por_mes.plot(kind='bar', color='skyblue')

# Añadir títulos y etiquetas para que el gráfico sea claro
plt.title('Cantidad Total de Pasajeros por Mes', fontsize=16)
plt.xlabel('Mes', fontsize=14)
plt.ylabel('Cantidad de Pasajeros', fontsize=14)
plt.xticks(rotation=45) # Rota las etiquetas del eje X para que no se superpongan
plt.grid(axis='y', linestyle='--', alpha=0.7) # Añade una cuadrícula suave en el eje Y

for etiqueta in etiquetas.patches:
    # 'etiqueta.get_height()' obtiene la altura de la barra (el valor)
    # 'etiqueta.get_x() + bar.get_width() / 2' calcula la posición X central de la barra
    # 'etiqueta.get_height()' + un pequeño offset para que el texto esté justo encima
    plt.text(etiqueta.get_x() + etiqueta.get_width() / 2, # Posición X del texto
             etiqueta.get_height() + (pasajeros_por_mes.max() * 0.02), # Posición Y del texto (un poco por encima de la barra)
             f'{int(etiqueta.get_height()):,}', # El valor a mostrar, formateado como entero con miles
             ha='center', # Alineación horizontal: centrado sobre la barra
             va='bottom', # Alineación vertical: la parte inferior del texto estará en la posición Y
             fontsize=10,
             color='dimgray')

plt.tight_layout() # Ajusta el diseño para evitar recortes

# Mostrar el gráfico
plt.show()
```

**La interpretación de tu gráfico anterior:**
Desde el año 1949 hasta el año 1960, se puede evidenciar que la cantidad de pasajeros ha ido aumentando año tras año.
Adicionalmente, los meses de julio y agosto son los meses que concentran mayor cantidad de pasajeros durante ese periodo de tiempo.
Mientras que el mes de febrero y noviembre son los meses que menor cantidad de pasajeros ha habido durante el mismo periodo.
A pesar que el mes de diciembre la mayoria de gente viaja por año nuevo, se nota que la data, son los meses de julio y agosto donde la mayor cantidad de gente viaja.

### **Ejercicio 3**: Método gráfico multivariado para datos cuantitativos ✋


```python
#Tu código aquí

# Crear un diccionario de mapeo de meses
mapeo_meses = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4,
    'May': 5, 'Jun': 6, 'Jul': 7, 'Aug': 8,
    'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12 }

# Convertir la columna 'Mes' usando el mapeo
# Usamos .loc para asegurar que la asignación se haga correctamente
df.loc[:, 'Mes_Num'] = df['month'].map(mapeo_meses)

plt.figure(figsize=(14, 10)) # Ajusta el tamaño para mejor visualización
sns.lineplot(data=df, x='Mes_Num', y='passengers', hue='year', marker='o', palette='viridis', lw=2)

# Configurar el eje X para mostrar nombres de meses
nombres_meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
# Asegurarse de que los ticks y etiquetas cubran los meses presentes en los datos
meses_presentes = sorted(df['Mes_Num'].unique())
plt.xticks(ticks=meses_presentes, labels=[nombres_meses[m-1] for m in meses_presentes])


plt.title('Cantidad de Pasajeros por Mes y Año', fontsize=18)
plt.xlabel('Mes', fontsize=15)
plt.ylabel('Cantidad de Pasajeros', fontsize=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Año', bbox_to_anchor=(1.05, 1), loc='upper left') # Mueve la leyenda fuera del gráfico
plt.tight_layout()
plt.show()
```

```python
# Pivotar el DataFrame: Años como columnas, Meses como filas, Valores como Cantidad de pasajeros
# Usamos .sort_values() para asegurar que los años estén ordenados en las columnas
df_pivot = df.pivot_table(index='Mes_Num', columns='year', values='passengers', aggfunc='sum').sort_values(by='Mes_Num', ascending=True)

plt.figure(figsize=(12, 8))
sns.heatmap(df_pivot, annot=True, fmt='d', cmap='YlGnBu', linewidths=.5, linecolor='white') # fmt='d' para enteros

# Configurar el eje Y para mostrar nombres de meses
nombres_meses_completos = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
plt.yticks(ticks=[i + 0.5 for i in range(len(df_pivot.index))], labels=[nombres_meses_completos[m-1] for m in df_pivot.index], rotation=0)

plt.title('Mapa de Calor de Pasajeros por Mes y Año', fontsize=18)
plt.xlabel('Año', fontsize=15)
plt.ylabel('Mes', fontsize=15)
plt.tight_layout()
plt.show()
```

**La interpretación de tu gráfico anterior:** ------
A través de ambos gráficos se valida que desde el año 1949 hasta el año 1960, los meses de julio y agosto son los meses con mayor cantidad de pasajeros.
Sin embargo con este último gráfico, podemos ver como enero y noviembre en los primeros años (1949 y 1950) eran los meses con menor cantidad de pasajeros; pero con el transcurrir de los años, enero fue incrementando y ya es un mes promedio de cantidad de pasajeros; mientras que noviembre si se mantiene como un mes bajo. Y esto tiene sentido dado que en diciembre y enero la gente viaja por fines de año y celebrar el año nuevo.
